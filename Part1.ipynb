```python
import cv2
import matplotlib.pyplot as plt

FD = cv2.CascadeClassifier('/content/AML-Project-Part1/haarcascade_frontalface_default.xml')
```

```python
# Function to take an image, perform Face detection, and then blur the face(s) present in the image
def blurFaces(img):
    # Detecting the face in an image using opencv's support via haarcascade
    faces = FD.detectMultiScale(img, scaleFactor=1.1, minNeighbors=4)

    # On each of the detected faces in the given image, performing Gaussian Blur
    for x, y, w, h in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), (255, 255, 255), 2)
        face = img[y:y + h, x:x + w]
        height, width = face.shape[0], face.shape[1]
        img[y:y + height, x:x + width] = cv2.GaussianBlur(face, (121, 121), 30)

    # Displaying the blurred image
    # Alternatively, could store this new image in a new file  
    plt.axis('off')
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
    plt.show()
```

```python
file = '/content/AML-Project-Part1/Humans/1 (1029).jpg'
image = cv2.cvtColor(cv2.imread(file), cv2.IMREAD_GRAYSCALE)
plt.axis('off')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
plt.show()
```

```python
blurFaces(image)
```
